```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([2018,2019,2020,2021,2022])
y = np.array([1050,2452,3456,4653,4052])

plt.plot(x,y)
plt.xticks(x)
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```

```python
plt.bar(x,y)
plt.xticks(x)
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```

```python
import pandas as pd
```

```python
df = pd.read_csv('datasets/gdp.csv')
df
```

```python
plt.plot(df['year'],df['gdp'])
plt.xticks(df['year'])
plt.title('Thailand GDP')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.show()
```

```python
plt.bar(
    df['year'], df['gdp'],
    color=['green','red','blue','yellow','pink'],
    edgecolor='black'
)
plt.xticks(df['year'])
plt.title('Thailand GDP')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.show()
```

```python
plt.pie(
    df['gdp'],
    labels=df['year'],
    autopct='%.2f%%',
    colors=['green','red','blue','yellow','pink']
)
plt.title('Thailand GDP')
plt.show()
```
