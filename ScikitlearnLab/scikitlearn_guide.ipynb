# Load Dataset ใน Scikit-Learn

```python
from sklearn import datasets

dir(datasets)
```

```python
[data for data in dir(datasets) if data.startswith('load')]
```

## Load iris dataset

```python
iris_dataset = datasets.load_iris()
```

```python
iris_dataset
```

```python
import pandas as pd

df = pd.DataFrame(data=iris_dataset.data, columns=iris_dataset.feature_names)
df
```

## แบ่งข้อมูล Train และ Test ของ IRIS

```python
from sklearn.datasets import load_iris

# function train_test_split
from sklearn.model_selection import train_test_split

# แบ่งข้อมูลออกเป็น 2 ส่วน train และ test
X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'], iris_dataset['target'], random_state=101)

# ดูข้อมูลทั้งหมด 150 ตัว
print('x_train shape:{}'.format(X_train.shape))
print('x_test shape:{}'.format(X_test.shape))
print('y_train shape:{}'.format(y_train.shape))
print('y_test shape:{}'.format(y_test.shape))
```

## ตัวอย่างการแบ่งข้อมูล Digits Dataset

```python
from sklearn.datasets import load_digits
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# load digit จาก sklearn
```

```python
digits = load_digits()

print('Image Data Shape', digits.data.shape)
```

```python
# โหลดข้อมูลเข้า Pandas Dataframe

df = pd.DataFrame(data=digits.data, columns=digits.feature_names)
df
```

```python
X_train, X_test, y_train, y_test = train_test_split(digits['data'], digits['target'], random_state=101)

print('x_train shape:{}'.format(X_train.shape))
print('x_test shape:{}'.format(X_test.shape))
print('y_train shape:{}'.format(y_train.shape))
print('y_test shape:{}'.format(y_test.shape))
```
